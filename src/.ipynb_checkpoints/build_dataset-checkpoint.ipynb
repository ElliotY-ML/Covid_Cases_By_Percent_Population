```python
# State Abbreviations
state_abbrev = []
with open(os.path.join('..','datasets','state_abbrev.txt')) as f:
    for i,l in enumerate(f.readlines()):
        state_abbrev.append(','.join(l.strip('\n').split('|')))
        
f.close()
```

```python
con.execute('DROP TABLE IF EXISTS state_abb')

con.execute('CREATE TABLE state_abb (STATE INTEGER,STUSAB text,STATE_NAME text,STATENS INTEGER)')

state_abb_list = []
for i in state_abbrev:
    state_abb_list.append(tuple(i.split(',')))

con.executemany('INSERT INTO state_abb VALUES (?,?,?,?)', state_abb_list[1:])

con.commit()
```
